## Ejercicio 2

La ecuación logística es un modelo simple para el crecimiento de una población $P$, dada su tasa de crecimiento $r$ y la capacidad de carga del ambiente $K$. Puede expresarse del siguiente modo:

$$\dot{P} = rP\left(1 - \frac{P}{K}\right).$$

### 2. a

Resolver la ecuación para distintos datos iniciales e interpretar el resultado obtenido.

```python
def euler_logistica(P0, h, r, K, T):
    P = P0
    t = 0
    lista = [P]
    while t < T:
        t = t + h
        P = P + r * P * (1 - P/K) * h
        lista.append(P)
    return lista 
```

### 2.d
Para estudiar el orden de convergencia del método, escribir una rutina que resuelva para distintos valores de $h$ (por ejemplo $h_m = 2^{-m}$ con $m = 0, \ldots, M$) y para cada uno de estos valores almacene el error en el último paso: $|y_N - y(5)|$ en un vector $\boldsymbol{e} = (e_0, \ldots, e_M)$. Observar que si suponemos que $e_i \sim kh_m^r$, entonces $\log(e_m) \sim c + r\log(h_i)$; por lo tanto, el gráfico de $\log(\boldsymbol{e})$ en función de $\log(\boldsymbol{h})$ debe ser una lineal de pendiente $r$. Realizar este gráfico. Operando con los vectores $\log(\boldsymbol{e})$ y $\log(\boldsymbol{h})$ estimar numéricamente el valor de $r$. ¿Qué se observa?

```python
def error_euler_logistica(P0, r, K, T, M):
    valor_exacto = K/(1 + ((K-P0)/P0) * np.exp(-r*T))
    resultados_finales = []    
    for i in range(1, M+1):
        P = P0
        t = 0
        h = 1/(2**i)
        while t < T:
            t = t + h
            P = P + r * P * (1 - P/K) * h
        resultados_finales.append(P)
    errores = [abs(x - valor_exacto) for x in resultados_finales]   
    return errores
```

```python
P0 = 2
r = 1
K = 100
errores = error_euler_logistica(P0, r, K, T, 7)
log_h = [np.log(1/2**x) for x in range(1,8)]
plt.xlabel("$\log(h)$")
plt.ylabel("$\log(e)$")
plt.plot(log_h,np.log(errores))
```

## Ejercicio 3

El nivel de líquido en un tanque de área transversal $A$ y el caudal de salida del mismo están dados por las siguientes ecuaciones:

$$\frac{dh(t)}{dt} = \frac{1}{A}\left(q_I - q_O\right)$$

$$h(0) = h_0$$

$$q_O = \frac{1}{R}\sqrt{h(t)}$$

donde $h$ es la altura de la columna de agua, $q_I$ y $q_O$ son los caudales de líquido de entrada y salida respectivamente y $R$ es la resistencia de salida.

### 3. a
Implementar el modelo asegurándose que $h(t) \geq 0$. Simular la evolución de la altura $h(t)$ y del caudal de salida $q_O(t)$ para $q_I = 1$, $A = 1$, $R = 0{,}5$ y condición inicial $h_0$.

```python
def tanque_euler(A,R, q_I,h_0, delta_tiempo, T):
    q_0 = 1/R * np.sqrt(h_0)
    vector_q_0 = [q_0]
    h = h_0
    t = 0
    alturas = [h]
    while(t<T):
        t = t + delta_tiempo
        q_0 = 1/R * np.sqrt(h)
        vector_q_0.append(q_0)
        candidato_h = h + 1/A * (q_I-q_0) * delta_tiempo 
        if (candidato_h >= 0):
            h = h + 1/A * (q_I-q_0) * delta_tiempo
        else:
            h = 0
        alturas.append(h)
    for i in range(len(alturas)):
        print(f"h = {alturas[i]:.3f}    q_O = {vector_q_0[i]:.3f}")
```

```python
tanque_euler(1,0.5, 1,20, 1/16, 1)
```

### 3.b
Analizar cómo cambian el nivel de líquido y el caudal de salida para $R$: $0{,}25$, $0{,}5$ y $0{,}95$. ¿Qué función cumple R?

```python
print("R = 0,25")
tanque_euler(1,0.25, 1,100, 1/16, 1)
print("R = 0,5")
tanque_euler(1,0.5, 1,100, 1/16, 1)
print("R = 0,95")
tanque_euler(1,0.95, 1,100, 1/16, 1)
```

A mayor resistencia, el caudal de salida es menor, por lo que la altura disminuye más lentamente.

```python
def f_velocidad(m,cr,v,t):
    g = 9.81
    return g - (cr/m * v**2)

def tiempos(h, T):
    N = int(T / h)
    t = np.linspace(0, T, N+1)
    return t
    
def euler_modif_ej_5(v0, cr,m,h, T):
    t = 0
    v = v0
    res = [v0]
    while t < T:
        v = v + h * f_velocidad(m, cr, v + h/2 * f_velocidad(m, cr,v,t),t + h/2)
        t = t + h
        res.append(v)
    return res
        
```

## Ejercicio 8

(Sistema predador-presa). Se tienen dos poblaciones, una de predadores y otra de presas, cuyo número a tiempo $t$ denotamos $x(t)$ e $y(t)$ respectivamente. La población de $x$ en ausencia de presas tiende a decaer a una tasa $\alpha$, mientras que la de $y$ en ausencia de predadores tiende a crecer a una tasa $\beta$. Por otra parte, los encuentros de predadores y presas hacen crecer la población de los primeros y decrecer la de los segundos, de acuerdo a cierta proporción. De este modo, se obtiene el sistema:

$$\dot{x} = -\alpha x + \gamma xy$$
$$\dot{y} = \beta y - \delta xy,$$

donde $\gamma y$ representa la tasa de crecimiento de $x$ (mayor cuanto más presas haya) y $\delta x$ representa la tasa de mortandad de presas (mayor cuanto más predadores haya). Se asume que los parámetros $\alpha, \beta, \gamma, \delta$ son todos positivos.

(a) Observar que el sistema tiene un equilibrio en el origen y otro equilibrio en $\left(\frac{\beta}{\delta}, \frac{\alpha}{\gamma}\right)$.

(b) Para $\alpha = \beta = \gamma = \delta = 0{,}5$, resolver usando la librería `DifferentialEquations.jl`, para distintos valores iniciales (cercanos y no tanto a los equilibrios). Para ello implementar una función `predpresa(u,p,t)` que devuelva `du`, la derivada de `u` en `t`. `p` es un vector de parámetros que puede usarse (o no), para darle a la función los valores de $\alpha, \beta, \gamma$ y $\delta$.

(c) Tomar $\alpha = 0{,}25$, $\beta = 1$, $\gamma = \delta = 0{,}01$, $x_0 = 80$ e $y_0 = 30$. Resolver y realizar gráficos como los del ítem anterior.

```python
def predpresa(u,t,p):
    x,y = u
    α, β, γ, δ = p
    dxdt = - α * x + γ * x * y
    dydt = β * y - δ * x * y
    
    return [dxdt, dydt]

def campo(X, Y, p):
    α, β, γ, δ = p
    dxdt = -α * X + γ * X * Y
    dydt =  β * Y - δ * X * Y
    return dxdt, dydt

p = (0.5,0.5,0.5,0.5)
x = np.linspace(0, 10, 11)
y = np.linspace(0, 10, 11)

X, Y = np.meshgrid(x, y)
U, V = campo(X, Y, p)
```
